```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
theta = np.arange(0,2*np.pi,0.1)
X = np.sin(theta)
Y = np.cos(theta)
Z = np.zeros(np.shape(theta))

xyz = np.vstack((X,Y,Z)).T

Yaw = 65
csY = np.array([[np.cos(Yaw*np.pi/180),np.sin(Yaw*np.pi/180),0],
               [-np.sin(Yaw*np.pi/180),np.cos(Yaw*np.pi/180),0],
       [0,0,1]])
xyzTheta = np.matmul(xyz,csY)

Roll = 20
csR = np.array([[np.cos(Roll*np.pi/180),0,np.sin(Roll*np.pi/180)],
                [0,1,0],
               [-np.sin(Roll*np.pi/180),0,np.cos(Roll*np.pi/180)]])
xyzRoll = np.matmul(xyzTheta,csR)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot(xyz[:,0],xyz[:,1],xyz[:,2],'.b')
ax.plot(xyz[0,0],xyz[0,1],xyz[0,2],'*r')
ax.plot(xyzTheta[:,0],xyzTheta[:,1],xyzTheta[:,2],'.g')
ax.plot(xyzTheta[0,0],xyzTheta[0,1],xyzTheta[0,2],'*m')
ax.plot(xyzRoll[:,0],xyzRoll[:,1],xyzRoll[:,2],'.k')
ax.plot(xyzRoll[0,0],xyzRoll[0,1],xyzRoll[0,2],'*y')
#ax.set(xticklabels=[],
#       yticklabels=[],
#       zticklabels=[])

plt.show()
```
