```python
    # number of bins in r and theta dimensions
    N_bins_theta = 90
    N_bins_r = 46

    # limits in r dimension
    rmin = 0
    rmax = 90

    # setting up 1D arrays in r and theta
    r = np.linspace(rmin, rmax, N_bins_r)
    theta = np.linspace(0, 2*np.pi, N_bins_theta)  # N.B. radians not degrees

    # 'gridding' the 1D arrays into 2D arrays so they can be used by pcolor
    theta, r = np.meshgrid(theta, r)
```

```python
    # generating random data; note shape of array
    H = np.linspace(0,2*np.pi,N_bins_theta)*np.ones([N_bins_r, N_bins_theta])
    V = np.full([N_bins_theta,N_bins_r],np.linspace(0,2*np.pi,N_bins_r)).transpose()

    shift = 20
    C = np.roll(H,shift,1) + np.roll(V,shift,0)
    
    # setting up 'polar' projection and plotting
    ax = plt.subplot(111, projection='polar')
    ax.set_theta_zero_location("N")
    ax.set_theta_direction(-1)
    ax.set_rlim([90,0])
    plt.grid(False)
    plt.pcolor(theta, r, C)
    plt.colorbar()
    plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
%matplotlib notebook


plt.style.use('_mpl-gallery')

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = 5-np.sqrt( X**2 + Y**2)
Z = (R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, R, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
