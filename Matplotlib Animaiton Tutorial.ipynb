## Standard Imports

```python
# Standard imports
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

## Setup Initial Plot

```python
# Set up empty Figure, Axes and Line objects
fig, ax = plt.subplots()
# Set axes limits so that the whole image is included
ax.set(xlim=(-0.1, 2*np.pi+0.1), ylim=(-1.1, 1.1))
# Draw a blank line
line, = ax.plot([], [])  
```

## Generate Sine Wave Data

```python
# Define data - one sine wave
x = np.linspace(0, 2*np.pi, num=50)
y = np.sin(x)
```

## Define Animation Function

Suttle point on "line," - The comma is required to returns a tuple so it is then iterable

```python
# Define animate function
def animate(i):
    line.set_data(x[:i], y[:i])
    return line,
```

## Pass to FuncAnimation
### Notes: 
    - additional +1 to frames for last point
    - interval is time (ms) between drawing next frame
    - blit only updates new/changed data and must be iterable (line,)
    - blit=FALSE: redraw complete plot each frame

```python
# Pass to FuncAnimation
anim = FuncAnimation(fig, animate, frames=len(x)+1, interval=30, blit=True)
```

# Visualize Animation in Jupyter Notebook

## By default [%matplotlib inline] is set.

The animation will not visualize and only spit back a handle

```python
# Standard imports
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Set up empty Figure, Axes and Line objects
fig, ax = plt.subplots()
# Set axes limits so that the whole image is included
ax.set(xlim=(-0.1, 2*np.pi+0.1), ylim=(-1.1, 1.1))
# Draw a blank line
line, = ax.plot([], []) 

# Define data - one sine wave
x = np.linspace(0, 2*np.pi, num=50)
y = np.sin(x)

# Define animate function
def animate(i):
    line.set_data(x[:i], y[:i])
    return line,

# Pass to FuncAnimation
anim = FuncAnimation(fig, animate, frames=len(x)+1, interval=30, blit=True)
```

## If we switch to [%matplotlib inline]

Must reset kernel before running!

```python
# Standard imports
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook

# Set up empty Figure, Axes and Line objects
fig, ax = plt.subplots()
# Set axes limits so that the whole image is included
ax.set(xlim=(-0.1, 2*np.pi+0.1), ylim=(-1.1, 1.1))
# Draw a blank line
line, = ax.plot([], []) 

# Define data - one sine wave
x = np.linspace(0, 2*np.pi, num=50)
y = np.sin(x)

# Define animate function
def animate(i):
    line.set_data(x[:i], y[:i])
    return line,

# Pass to FuncAnimation
anim = FuncAnimation(fig, animate, frames=len(x)+1, interval=30, blit=True)
```

Just called before anim!
